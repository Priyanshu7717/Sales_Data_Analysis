```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/sales_data_10000.csv")
```

```python

# Check missing values in each column
df.isnull().sum()

# Check duplicate records
df.duplicated().sum()

# Display data types of all columns
df.dtypes

# Count unique values in each column
df.nunique()
```

```python
# ==========================================================
# Step 1 : State-wise Sales Analysis
# Purpose : Analyze total sales across different states
# ==========================================================

# Calculate state-wise total sales
state_sales = df.groupby("State")["Sales"].sum()

# Display state-wise sales
print("========== State-wise Sales ==========\n")
print(state_sales)

# Display states with highest sales
print("\n========== Top 10 States by Sales ==========\n")
print(state_sales.sort_values(ascending=False).head(10))

# Display states with lowest sales
print("\n========== Bottom 10 States by Sales ==========\n")
print(state_sales.sort_values().head(10))
```

```python
# ==========================================================
# Step 2 : State-wise Sales Visualization
# ==========================================================

# Create figure
plt.figure(figsize=(14,6))

# Plot state-wise sales
state_sales.sort_values(ascending=False).plot(kind="bar")

# Add chart title
plt.title("State-wise Sales")

# Add axis labels
plt.xlabel("State")
plt.ylabel("Sales")

# Display chart
plt.show()
```

```python
# ==========================================================
# Step 3 : Category-wise Sales Analysis
# ==========================================================

# Calculate category-wise total sales
category_sales = df.groupby("Category")["Sales"].sum()

# Display category-wise sales
print("========== Category-wise Sales ==========\n")
print(category_sales)

# Display categories sorted by sales
print("\n========== Categories Ranked by Sales ==========\n")
print(category_sales.sort_values(ascending=False))
```

```python
# ==========================================================
# Step 3 : Category-wise Sales Visualization
# ==========================================================

# Create figure
plt.figure(figsize=(8,5))

# Plot category-wise sales
category_sales.sort_values(ascending=False).plot(kind="bar")

# Add chart title
plt.title("Category-wise Sales")

# Add axis labels
plt.xlabel("Category")
plt.ylabel("Sales")

# Display grid
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Save chart
plt.savefig("../charts/category_sales.png", dpi=300, bbox_inches="tight")
# Display chart
plt.show()
```

```python
# ==========================================================
# Step 4 : Product-wise Sales Analysis
# ==========================================================

# Calculate product-wise total sales
product_sales = df.groupby("Product")["Sales"].sum()

# Display product-wise sales
print("========== Product-wise Sales ==========\n")
print(product_sales)

# Display top 10 selling products
print("\n========== Top 10 Selling Products ==========\n")
print(product_sales.sort_values(ascending=False).head(10))
```

```python
# ==========================================================
# Step 4 : Top 10 Products Visualization
# ==========================================================

# Get top 10 products
top_products = product_sales.sort_values(ascending=False).head(10)

# Create figure
plt.figure(figsize=(10,6))

# Plot horizontal bar chart
plt.barh(top_products.index, top_products.values)

# Add chart title
plt.title("Top 10 Selling Products")

# Add axis labels
plt.xlabel("Sales")
plt.ylabel("Product")

# Display grid
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Save Chart
plt.savefig("../charts/product_sales.png", dpi=300, bbox_inches="tight")

# Display chart
plt.show()
```

```python
# ==========================================================
# Step 5 : Monthly Sales Trend Analysis
# ==========================================================

# Convert Order_Date into datetime format
df["Order_Date"] = pd.to_datetime(df["Order_Date"])

# Extract month name
df["Month"] = df["Order_Date"].dt.month_name()

# Calculate monthly sales
monthly_sales = df.groupby("Month")["Sales"].sum()

# Display monthly sales
print("========== Monthly Sales ==========\n")
print(monthly_sales)
```

```python
# ==========================================================
# Step 5 : Monthly Sales Trend Visualization
# ==========================================================

# Arrange months in calendar order
month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_sales = monthly_sales.reindex(month_order)

# Create figure
plt.figure(figsize=(12,5))

# Plot line chart
plt.plot(monthly_sales.index, monthly_sales.values, marker="o")

# Add chart title
plt.title("Monthly Sales Trend")

# Add axis labels
plt.xlabel("Month")
plt.ylabel("Sales")

# Display grid
plt.grid(True, linestyle="--", alpha=0.7)

# Save Chart
plt.savefig("../charts/monthly_sales.png", dpi=300, bbox_inches="tight")

# Display chart
plt.show()
```

```python
# ==========================================================
# Step 6 : Payment Mode Analysis
# ==========================================================

# Count orders by payment mode
payment_mode = df["Payment_Mode"].value_counts()

# Display payment mode count
print("========== Payment Mode Count ==========\n")
print(payment_mode)

# Calculate payment mode percentage
payment_percentage = (payment_mode / payment_mode.sum()) * 100

# Display payment mode percentage
print("\n========== Payment Mode Percentage ==========\n")
print(payment_percentage.round(2).astype(str) + "%")
```

```python
# ==========================================================
# Step 6 : Payment Mode Visualization
# ==========================================================

# Create figure
plt.figure(figsize=(7,7))

# Plot payment mode distribution
plt.pie(
    payment_mode,
    labels=payment_mode.index,
    autopct="%1.1f%%",
    startangle=90
)

# Add chart title
plt.title("Payment Mode Distribution")

# Save Chart
plt.savefig("../charts/payment_mode.png", dpi=300, bbox_inches="tight")

# Display chart
plt.show()
```

```python
# ==========================================================
# Step 7 : Region-wise Sales Analysis
# ==========================================================

# Calculate region-wise total sales
region_sales = df.groupby("Region")["Sales"].sum()

# Display region-wise sales
print("========== Region-wise Sales ==========\n")
print(region_sales)

# Display regions sorted by sales
print("\n========== Regions Ranked by Sales ==========\n")
print(region_sales.sort_values(ascending=False))
```

```python
# ==========================================================
# Step 7 : Region-wise Sales Visualization
# ==========================================================

# Create figure
plt.figure(figsize=(8,5))

# Plot region-wise sales
plt.bar(region_sales.index, region_sales.values)

# Add chart title
plt.title("Region-wise Sales")

# Add axis labels
plt.xlabel("Region")
plt.ylabel("Sales")

# Display grid
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Save Chart
plt.savefig("../charts/region_sales.png", dpi=300, bbox_inches="tight")

# Display chart
plt.show()
```

```python
# ==========================================================
# Step 8 : Category-wise Profit Analysis
# ==========================================================

# Calculate category-wise total profit
category_profit = df.groupby("Category")["Profit"].sum()

# Display category-wise profit
print("========== Category-wise Profit ==========\n")
print(category_profit)

# Display categories ranked by profit
print("\n========== Categories Ranked by Profit ==========\n")
print(category_profit.sort_values(ascending=False))
```

```python
# ==========================================================
# Step 8 : Category-wise Profit Visualization
# ==========================================================

# Create figure
plt.figure(figsize=(8,5))

# Plot category-wise profit
plt.bar(category_profit.index, category_profit.values)

# Add chart title
plt.title("Category-wise Profit")

# Add axis labels
plt.xlabel("Category")
plt.ylabel("Profit")

# Display grid
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Display chart
plt.show()
```

```python
# ==========================================================
# Step 9 : Customer Analysis
# ==========================================================

# Calculate customer-wise total sales
customer_sales = df.groupby("Customer_Name")["Sales"].sum()

# Display top 10 customers
print("========== Top 10 Customers ==========\n")
print(customer_sales.sort_values(ascending=False).head(10))
```

```python
# ==========================================================
# Step 9 : Top 10 Customers Visualization
# ==========================================================

# Get top 10 customers
top_customers = customer_sales.sort_values(ascending=False).head(10)

# Create figure
plt.figure(figsize=(10,6))

# Plot horizontal bar chart
plt.barh(top_customers.index, top_customers.values)

# Add chart title
plt.title("Top 10 Customers by Sales")

# Add axis labels
plt.xlabel("Sales")
plt.ylabel("Customer Name")

# Display grid
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Save Chart
plt.savefig("../charts/top_customers.png", dpi=300, bbox_inches="tight")
# Display chart
plt.show()
```
